# EDA con pandas

Ahora haremos graficas utilizando matplotlib y nos poyaremos usando pandas para completar informacion necesaria.

```python
import pandas as pd
from matplotlib import pyplot as plt
df = pd.read_csv('Limpieza_final_final_ahora_si.csv')
df.head()
```

```python
#Cuál es la distribución de los salarios en la empresa
plt.hist(df['salary'], bins=40, edgecolor='darkblue')
plt.title('Distribución de los salarios en la empresa')
plt.xlabel('Salario')
plt.ylabel('Frecuencia')
plt.show()
```

```python
#Cuál es la proporción de empleados por género
df['genre'].value_counts().plot(kind='bar', color='green')
plt.title('Proporción de empleados por género')
plt.xlabel("Genero")
plt.ylabel("Total De Empleados")   
plt.show()
```

```python
#¿Cómo varían los salarios entre diferentes áreas de la empresa?
df.groupby('area')['salary'].mean().sort_values(ascending=True).plot(kind='bar', color='green')
plt.title('Salarios Promedio por Área')
plt.xlabel('Área')
plt.ylabel('Salario Promedio')
plt.show()
```

```python
#Distribucion de edades

""" plt.hist(df['age'], bins=40, edgecolor='darkblue')
plt.title('Distribución de edades en la empresa')
plt.xlabel('Edades')
plt.ylabel('Frecuencia')
plt.show() """

df['age'].value_counts().sort_values(ascending=True).plot(kind='bar', color='blue')
plt.title('Distribución de edades en la empresa')
plt.xlabel("Edad")
plt.ylabel("Total De Empleados")   
plt.show()
```

```python
#Salario por edad 
df.groupby('age')['salary'].mean().sort_values(ascending=True).plot(kind='bar', color='green')
plt.title('Salarios Promedio por Edad')
plt.xlabel('Edad')
plt.ylabel('Salario Promedio')
plt.show()
```

```python
#Cuantos tipos de empleados hay
df['type'].value_counts().plot(kind='bar', color='purple', grid=True)
plt.title('Proporción de empleados por tipo')
plt.xlabel("Tipo")
plt.ylabel("Total De Empleados")   
plt.show()
```

```python
#Ahora crearemos una columna del año en el que empezaron a trabajar a partir de la columna de start_year
df['start_date'] = pd.to_datetime(df["start_date"], yearfirst=True)#4
df['start_yeart'] = df['start_date'].dt.year
df.head()
```

```python
#Grafica de la proporcion de empleados por el año en el que ingresaron a la empresa
df['start_yeart'].value_counts().sort_values(ascending=True).plot(kind='bar', color='green')
plt.title('Proporción de empleados por año en el que empezaron a trabajar')
plt.xlabel("Año")
plt.ylabel("Total De Empleados")   
plt.show()
```
